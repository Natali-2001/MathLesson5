```python
#Задание 2.1. Написать код, проверяющий любую из теорем сложения или умножения вероятностей.
%matplotlib inline
import numpy as np
orel = 0
reshka = 0
n = 100
for i in range(0,n):
    x = np.random.uniform(0, 10)
    if x<5:
        orel = orel + 1
    else:
        reshka = reshka + 1
print("Решек", reshka)
print("Орлов", orel)
print(f"Вероятность выпадения орла или решки {orel / n + reshka / n}, что равно 1 - сумме вероятностей несовместных событий")
```

```python
#Задание 3.1. Дополните код Монте-Карло... 
import numpy as np
import itertools
import math
k, n = 0, 10000
a = np.random.randint(0, 2, n)
b = np.random.randint(0, 2, n)
c = np.random.randint(0, 2, n)
d = np.random.randint(0, 2, n)
x = a + b + c + d
for i in range(0, n):
    if x[i] == 2:
        k = k + 1      
#биномиальное распределение
P = math.factorial(4)/(math.factorial(2)*math.factorial(4-2)) * (1/2)**4
#Разница расчетов между методом Монте-Карло и расчетом вероятностей
y = round(abs((1 - (k/n) / P) * 100),2)
print(f"Вероятность выпадения 2 успехов на 4 испытаниях по методу Монте-Карло: {k/n}")
print(f"Вероятность выпадения 2 успехов на 4 испытаниях при биномиальном распределении: {P}")
print(f"Разница расчетов между методом Монте-Карло и расчетом вероятностей составила: {y} %")
```

```python
#Задание 3.2. Повторите расчеты биномиальных коэффициентов..
import numpy as np
import itertools
import math
k, n = 0, 1000
a = np.random.randint(0, 2, n)
b = np.random.randint(0, 2, n)
c = np.random.randint(0, 2, n)
d = np.random.randint(0, 2, n)
x = a + b + c + d
for i in range(0, n):
    if x[i] == 3:
        k = k + 1      
#биномиальное распределение
P = math.factorial(4)/(math.factorial(3)*math.factorial(4-3)) * (1/2)**3 * (1/2)**1
#Разница расчетов между методом Монте-Карло и расчетом вероятностей
y = round(abs((1 - (k/n) / P) * 100),2)
print(f"Вероятность выпадения 3 успехов на 4 испытаниях по методу Монте-Карло: {k/n}")
print(f"Вероятность выпадения 3 успехов на 4 испытаниях при биномиальном распределении: {P}")
print(f"Разница расчетов между методом Монте-Карло и расчетом вероятностей составила: {y} %")
```

```python
# Задание 5. Дополните код расчетом коэффициентов корреляции x и y по формуле.
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
n = 100
r = 0.7
x = np.random.rand(n)
y = r*x + (1 - r)*np.random.rand(n)
plt.plot(x, y, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
a = (np.sum(x)*np.sum(y) - n*np.sum(x*y))/(np.sum(x)*np.sum(x) - n*np.sum(x*x))
b = (np.sum(y) - a*np.sum(x))/n
xm = np.sum(x)/n
ym = np.sum(y)/n
R = (np.sum((x-xm)*(y-ym))/np.sqrt(np.sum((x-xm)**2 * np.sum((y-ym)**2))))
A = np.vstack([x, np.ones(len(x))]).T
a1, b1 = np.linalg.lstsq(A, y)[0]
print(a, b)
print(a1, b1)
print(f"Коэффициент корреляции равен {R}")
plt.plot([0, 1], [b, a + b])
plt.show()
#проверка правильности вычисления коэффициента корреляции
c = np.corrcoef(x, y)
print(c)
```
